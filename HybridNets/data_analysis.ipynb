```python
import numpy 
import pandas
import json
```

```python
path = "/home/aman/Projects/Auto/Autonomous-Bicycle/HybridNets/logs/bdd100k_nrp_a100/tensorboard/metrics.jsonl"

train_metrics = []
val_metrics = []

with open(path) as f:
   for i, line in enumerate(f):
        line = line.strip()
        if line:
            if i % 2 == 0:
                train_metrics.append(json.loads(line))
            else:
                val_metrics.append(json.loads(line))
```

```python
train_metrics_df = pandas.DataFrame(train_metrics)
val_metrics_df = pandas.DataFrame(val_metrics)
```
