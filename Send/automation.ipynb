```python
def run_model_and_overlay(frame, image_numpy):
    overlay = frame.copy()
    # print(f"overlay shape:", overlay.shape)
    # print(overlay)
    

    overlay[image_numpy == 1] = (0, 255, 0)      # road
    overlay[image_numpy == 2] = (0, 0, 255)      # lane edge

    result = cv2.addWeighted(
        frame,
        0.7,      # original image weight
        overlay,
        0.3,      # mask weight
        0
    )
    return result
    
```

```python
video_width = 1080
video_hiegh = 1920
output_height = 384
output_width = 640
fps = 30
output_path = "numpy_concat/numpy_videos_concat2.mp4"


writer = cv2.VideoWriter(
    output_path,
    cv2.VideoWriter_fourcc(*'mp4v'),
    fps,
    (output_width, output_height)
)


print(writer)

if not writer.isOpened():
      raise RuntimeError(f"Could not open writer: {output_path}")

i = 0
while i < 1000:#total_frames/2:
    
    ok, frame_bgr = video_file.read()
    if not ok:
        print(f"Error")
        
        continue

    
    input_array, output_array = preprocess_frame(
            frame_bgr,
            video_width,
            video_hiegh,
            output_width,
            output_height,
        )
  
    image_numpy = numpy_file[i]
    result = run_model_and_overlay(output_array,image_numpy)
    if result.shape[:2] != (384, 640):
          result = cv2.resize(result, (640, 384))
            
    writer.write(result)
    i += 1
    
    if i % 900 == 0:
        print(f"Processed {i} frames")
    
print("Releasing resources...")
writer.release()

```
